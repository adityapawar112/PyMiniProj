```python
class EnergySystem:
    def __init__(self, building_name, energy_consumtion, emission_factor):
        self.building_name = building_name
        self.energy_consumtion = energy_consumtion
        self.emission_factor = emission_factor
    
    def calculate_carbon_footprint(self):
        return self.energy_consumtion * self.emission_factor
    
    def calculate_energy_savings(self):
        return self.energy_consumtion * 0.10
    

building = EnergySystem("Green Building", 15000, 0.2)
print(f"Building Name: {building.building_name}")
print(f"Energy Consumption: {building.energy_consumtion} kWh")
print(f"Emission Factor: {building.emission_factor} kg CO2/kWh")
print(f"Carbon Footprint: {building.calculate_carbon_footprint()} kg CO2")
print(f"Energy Savings: {building.calculate_energy_savings()} kWh")
```

```python
class SolarEnergySystem(EnergySystem):
    def __init__(self, building_name, energy_consumtion, emission_factor, solar_production):
        super().__init__(building_name, energy_consumtion, emission_factor)
        self.solar_production = solar_production
    
    def calculate_carbon_footprint(self):  # Override parent method
        net_consumption = self.energy_consumtion - self.solar_production
        return net_consumption * self.emission_factor

    def net_energy_consumption(self):
        return self.energy_consumtion - self.solar_production

solar_building = SolarEnergySystem("Solar Building", 15000, 0.2, 5000)   
print(f"\nBuilding Name: {solar_building.building_name}")
print(f"Energy Consumption: {solar_building.energy_consumtion} kWh")
print(f"Emission Factor: {solar_building.emission_factor} kg CO2/kWh")
print(f"Solar Production: {solar_building.solar_production} kWh")
print(f"Net Energy Consumption: {solar_building.net_energy_consumption()} kWh")
print(f"Carbon Footprint: {solar_building.calculate_carbon_footprint()} kg CO2")
```

```python
class EncapsulatedEnergySystem:
    def __init__(self, building_name, energy_consumtion, emission_factor):
        self.building_name = building_name
        self._energy_consumtion = energy_consumtion
        self.__emission_factor = emission_factor
    def get_energy_consumtion(self):
        return self._energy_consumtion
    
    def calculate_carbon_footprint(self):
        return self._energy_consumtion * self.__emission_factor
    
encapsulated_building = EncapsulatedEnergySystem("Encapsulated Building", 15000, 0.2)
print(f"\nBuilding Name: {encapsulated_building.building_name}")
print(f"Energy Consumption: {encapsulated_building.get_energy_consumtion()} kWh")
print(f"Emission Factor: {encapsulated_building.calculate_carbon_footprint()} kg CO2")
```

```python
class Building:
    def __init__(self, name, energy_consumption):
        self.name = name
        self.energy_consumption = energy_consumption

    def calculate_carbon_footprint(self, energy_consumtion):
        return self.energy_consumption * 0.2
    
class ResidentialBuilding(Building):
    def __init__(self, name, energy_consumption, ):
        super().__init__(name, energy_consumption)

    def calculate_carbon_footprint(self):
        return self.energy_consumption * 0.15
    
class CommercialBuilding(Building):
    def __init__(self, name, energy_consumption):
        super().__init__(name, energy_consumption)

    def calculate_carbon_footprint(self):
        return self.energy_consumption * 0.25
    
class IndustrialBuilding(Building):
    def __init__(self, name, energy_consumption):
        super().__init__(name, energy_consumption)

    def calculate_carbon_footprint(self):
        return self.energy_consumption * 0.3

residential = ResidentialBuilding("Residential Building", 12000)
commercial = CommercialBuilding("Commercial Building", 20000)
industrial = IndustrialBuilding("Industrial Building", 50000)

print(f"\n{residential.name} Carbon Footprint: {residential.calculate_carbon_footprint()} kg CO2")
print(f"{commercial.name} Carbon Footprint: {commercial.calculate_carbon_footprint()} kg CO2")
print(f"{industrial.name} Carbon Footprint: {industrial.calculate_carbon_footprint()} kg CO2")
```

```python
class GreenHouse:
    def __init__(self, temperature, humidity):
        self._temperature = temperature
        self._humidity = humidity
    
    def get_temperature(self):
        return self._temperature
    
    def get_humidity(self):
        return self._humidity
    
    def set_temperature(self, temperature):
        self._temperature = temperature
    
    def set_humidity(self, humidity):
        self._humidity = humidity
        
greenhouse1 = GreenHouse(22, 60)
print(f"\nGreenhouse Temperature: {greenhouse1.get_temperature()} 째C")
print(f"Greenhouse Humidity: {greenhouse1.get_humidity()} %")
greenhouse1.set_temperature(25)
greenhouse1.set_humidity(65)
print(f"Updated Greenhouse Temperature: {greenhouse1.get_temperature()} 째C")
print(f"Updated Greenhouse Humidity: {greenhouse1.get_humidity()} %")
```

```python
class WaterConservation:
    def __init__(self, water_collected, water_used):
        self.water_collected = water_collected
        self.water_used = water_used
    
    def calculate_water_savings(self):
        return self.water_collected - self.water_used
    
class RainwaterHarvesting(WaterConservation):
    def __init__(self, water_collected, water_used):
        super().__init__(water_collected, water_used)
    
    def calculate_water_savings(self):
        return (self.water_collected - self.water_used) * 90

class DripIrrigation(WaterConservation):
    def __init__(self, water_collected, water_used):
        super().__init__(water_collected, water_used)
    
    def calculate_water_savings(self):
        return (self.water_collected - self.water_used) * 80
    
class WaterRecycling(WaterConservation):
    def __init__(self, water_collected, water_used):
        super().__init__(water_collected, water_used)
    
    def calculate_water_savings(self):
        return (self.water_collected - self.water_used) * 70
    
rainwater = RainwaterHarvesting(1000, 200)
drip_irrigation = DripIrrigation(800, 150)
water_recycling = WaterRecycling(600, 100)
print(f"\nRainwater Harvesting Water Savings: {rainwater.calculate_water_savings()} liters")
print(f"Drip Irrigation Water Savings: {drip_irrigation.calculate_water_savings()} liters")
print(f"Water Recycling Water Savings: {water_recycling.calculate_water_savings()} liters")
```

```python
from abc import ABC, abstractmethod

# Abstract Base Class(ABC) - This class will define what every energy source must be able to do, but not how it does it.
class EnergySource(ABC):
    @abstractmethod #Declares required methods
    def calculate_energy_output(self):
        pass

    @abstractmethod
    def calculate_co2_savings(self):
        pass
```

```python
class SolarPanel(EnergySystem):
    def __init__(self, hours_of_sunlight, efficiency):
        self.hours_of_sunlight = hours_of_sunlight
        self.efficiency = efficiency

    def calculate_energy_output(self):
        return self.hours_of_sunlight * self.efficiency
        
    def calculate_co2_savings(self):
        return self.calculate_energy_output() * 0.5

class WindTurbine(EnergySystem):
    def __init__(self, wind_speed, turbine_efficiency):
        self.wind_speed = wind_speed
        self.turbine_efficiency = turbine_efficiency

    def calculate_energy_output(self):
        return self.wind_speed * self.turbine_efficiency * 0.5
    
    def calculate_co2_savings(self):
        return self.calculate_energy_output() * 0.3

class HydroTurbine(EnergySystem):
    def __init__(self, water_flow_rate, turbine_efficiency):
        self.water_flow_rate = water_flow_rate
        self.turbine_efficiency = turbine_efficiency

    def calculate_energy_output(self):
        return self.water_flow_rate * self.turbine_efficiency * 0.8
    
    def calculate_co2_savings(self):
        return self.calculate_energy_output() * 0.4
        
solar_panel = SolarPanel(5, 0.15)
wind_turbine = WindTurbine(10, 0.2)
hydro_turbine = HydroTurbine(100, 0.25)
print(f"\nSolar Panel Energy Output: {solar_panel.calculate_energy_output()} kWh")
print(f"Solar Panel CO2 Savings: {solar_panel.calculate_co2_savings()} kg CO2")
print(f"Wind Turbine Energy Output: {wind_turbine.calculate_energy_output()} kWh")
print(f"Wind Turbine CO2 Savings: {wind_turbine.calculate_co2_savings()} kg CO2")
print(f"Hydro Turbine Energy Output: {hydro_turbine.calculate_energy_output()} kWh")
print(f"Hydro Turbine CO2 Savings: {hydro_turbine.calculate_co2_savings()} kg CO2")
```

```python
def total_co2_absorbed(years, co2_per_tree=20):
    """
    Recursive function to calculate total CO2 absorbed over n years.
    Each person plants 1 tree and inspires 2 more people next year.
    """
    if years == 0:
        return 0
    # Number of trees planted in this year = 2^(years-1)
    trees_this_year = 2 ** (years - 1)
    return (trees_this_year * co2_per_tree) + total_co2_absorbed(years - 1, co2_per_tree)

n = 5  # Number of years
print(f"Total CO2 absorbed over {n} years: {total_co2_absorbed(n)} kg")



    
```
