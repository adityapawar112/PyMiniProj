```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sklearn as sk
import seaborn as sns
from sklearn.datasets import load_iris
```

```python
iris= sns.load_dataset("iris")
print(iris.head())  # View first 5 rows
```

```python
# 1. Histogram of sepal_length
plt.figure(figsize=(6,4))
sns.histplot(iris['sepal_length'], kde=True)
plt.title("Histogram of Sepal Length")
plt.show()
```

```python
# 2. Boxplot of petal_length by species
plt.figure(figsize=(6,4))
sns.boxplot(x='species', y='petal_length', data=iris)
plt.title("Boxplot of Petal Length by Species")
plt.show()
```

```python
# 3. Violin plot of sepal_width by species
plt.figure(figsize=(6,4))
sns.violinplot(x='species', y='sepal_width', data=iris)
plt.title("Violin Plot of Sepal Width by Species")
plt.show()
```

```python
# 4. Scatter plot of sepal_length vs petal_length
plt.figure(figsize=(6,4))
sns.scatterplot(x='sepal_length', y='petal_length', hue='species', data=iris)
plt.title("Scatter Plot: Sepal Length vs Petal Length")
plt.show()
```

```python
# 5. Heatmap of feature correlations
plt.figure(figsize=(6,4))
corr = iris.drop('species', axis=1).corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
# Exclude the 'species' column before calculating correlation
correlation_matrix = iris.drop('species', axis=1).corr()

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidth=0.5)

plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
sns.boxplot(data=iris, x="species", y="sepal_length")
plt.title("Boxplot of Sepal Length by Species")
plt.show()
sns.boxplot(data=iris, x="species", y="sepal_width")
plt.title("Boxplot of Sepal Width by Species")
plt.show()
sns.boxplot(data=iris, x="species", y="petal_length")
plt.title("Boxplot of Petal Length by Species")
plt.show()
sns.boxplot(data=iris, x="species", y="petal_width")
plt.title("Boxplot of Petal Width by Species")
plt.show()
```
