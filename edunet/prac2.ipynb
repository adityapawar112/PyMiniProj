```python
climate_data = [
    {"city": "City A", "temperature": 25, "carbon_footprint": 500},
    {"city": "City B", "temperature": 30, "carbon_footprint": 350},
    {"city": "City C", "temperature": 22, "carbon_footprint": 600},
    {"city": "City D", "temperature": 15, "carbon_footprint": 200},
    {"city": "City E", "temperature": 28, "carbon_footprint": 450},
]

for i in climate_data:
    if i['temperature'] > 26:
        print(f"{i['city']} has a high temperature.")
    
```

```python
total = 0
count = 0
for i in climate_data:
    total += i['carbon_footprint']
    count += 1
average_cf = total / count
print(average_cf)
```

```python
cf_threshold = average_cf
for i in climate_data:
    if i['carbon_footprint'] > cf_threshold:
        i['is_sustainable'] = False
    else:
        i['is_sustainable'] = True
print(climate_data)
```

```python
def even_odd(n):
    if n % 2 == 0:
        return "even"
    else:
        return "odd"

input_number = int(input("Enter a number: "))
print(even_odd(input_number))
```

```python
def carbon_footprint_calculator(electricity_usage, transport_km, waste_produced, water_usage):

    electricity_factor = 0.5         
    transport_factor = 0.79          
    waste_factor = 2                 
    water_factor = 0.002             

    # Calculations (monthly basis)
    electricity_emissions = electricity_usage * electricity_factor
    transport_emissions = transport_km * 4 * transport_factor  
    waste_emissions = waste_produced * 4 * waste_factor
    water_emissions = water_usage * 30 * water_factor  

    total_emissions = electricity_emissions + transport_emissions + waste_emissions + water_emissions
    return total_emissions


electricity = float(input("Enter your monthly electricity usage (in kWh): "))
transport = float(input("Enter your weekly transport distance (in miles): "))
waste = float(input("Enter your weekly waste production (in kg): "))
water = float(input("Enter your daily water usage (in liters): "))
total = carbon_footprint_calculator(electricity, transport, waste, water)
print(f"\n Your estimated monthly carbon footprint is: {total:.2f} kg CO₂")
```

```python
def is_energy_efficient(appliance_rating):
    if appliance_rating >= 4:
        return "Energy Efficient"
    else:
        return "Not Energy Efficient"

rating = int(input("Enter the appliance rating (1-5): "))
if rating < 1 or rating > 5:
    print("Invalid rating. Please enter a number between 1 and 5.")
    rating = int(input("Enter the appliance rating (1-5): "))

print(is_energy_efficient(rating))
```

```python
def solar_output():
    total_output = 0
    for i in range(7):
        daily_output = float(input(f"Enter the solar output for day {i+1} (in kWh): "))
        if daily_output < 0:
            print("Invalid input. Solar output cannot be negative.")
            return
        total_output += daily_output
    print(f"Total solar output for the week: {total_output} kWh")

solar_output()
```

```python
def high_carbon_footprint():
    for i in range(7):
        daily_milage = float(input(f"Enter the mileage for day {i+1} (in km): "))
        if daily_milage < 50:
            print(f"Day {i+1} has low carbon footprint.")
        else:
            print(f"Day {i+1} has high carbon footprint.")

high_carbon_footprint()
```

```python
def rainwater_harvesting_calculator(rainfall, roof_area):
    if rainfall < 0 or roof_area <= 0:
        print("Invalid input. Rainfall must be non-negative and roof area must be positive.")
    
    collected_water = rainfall * roof_area * 0.9 
    print(f"Estimated water collected: {collected_water:.2f} liters")

rainwater_harvesting_calculator(float(input("Enter the rainfall in mm: ")), float(input("Enter the roof area in square meters: ")))
```

```python
def energy_consumption_analyzer():
    total_consumption = 0
    average_consumption = 0
    for i in range(7):
        daily_consumption = float(input(f"Enter the energy consumption for day {i+1} (in kWh): "))
        total_consumption += daily_consumption
    average_consumption = total_consumption / 7
    if average_consumption < 10:
        print("You are within green energy consumption limits.")
    else:
        print("You are exceeding green energy consumption limits.")

energy_consumption_analyzer()
```
